# Workshop

```python
x = [1,2,3,4,5,6,7,8]
y = [1,4,9,16,25,36,49,64]

x2 = [1, 2, 3, 4, 5, 6, 7, 8]
y2 = [1, 8, 27, 64, 125, 216, 343, 512]
```

```python
plt.plot(x, y, label = 'X Kare', color = '#ffee56') # label komutu ile legendleri belirledik
plt.plot(x2, y2, label = 'X Küp', color = 'green')  # İki veri setini üst üste yazdık.

plt.title('Line Plot') # Başlık
plt.xlabel('X Ekseni') # Etiket
plt.ylabel('Y Ekseni')

plt.legend()
plt.grid() # Gridleri açtık.
plt.show() 
```

# Subplots

```python
fig = plt.figure(figsize = (11, 5))

first_plot = fig.add_subplot(1,3,1) # 1 satır, 3 sütun, 1 grafiğin konumu.
first_plot.plot(x,y, color = 'red')
first_plot.set_title('Line Plot')
first_plot.grid()

second_plot = fig.add_subplot(1,3,2)
second_plot.scatter(x2,y2, color = 'green')
second_plot.set_title('Scatter Plot')
second_plot.grid()

third_plot = fig.add_subplot(1,3,3)
third_plot.bar(x,y, color = 'yellow')
third_plot.set_title('Bar Graph')
third_plot.grid()

plt.show()
```
